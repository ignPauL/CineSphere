```python
import numpy as np
import pandas as pd
import ast
```

```python
movies = pd.read_csv("tmdb_5000_movies.csv")
credits = pd.read_csv("tmdb_5000_credits.csv")
```

```python
movies.head()
```

```python
movies.head(1)
```

```python
credits.head(1)
```

```python
#First we will merge both the dataframes or csv files
movies = movies.merge(credits, on= 'title')
```

```python
movies.head()
```

```python
# Removing the useless columns from the dataset and those columns are not so important for recommendation of movies

# Important columns are : genres, id, keywords, title, overview, cast , crew

movies = movies[['movie_id','title', 'overview', 'genres', 'keywords', 'cast', 'crew']]
```

```python
movies.head()
```

```python
# Removing the missing data
movies.isnull().sum()
```

```python
# We can see that there are 3 missing data in overview so we have to remove them
movies.dropna(inplace=True)
```

```python
# Removing duplicate data
movies.duplicated().sum()
# As 0 is coming so there is no duplication
```

```python
movies.iloc[0].genres
# As we can see it is in a weird format. As we can see it is a string of lists where we want a single list so we have to use a module named "ast"
```

```python

def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python

def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if (counter!=3):
             L.append(i['name'])
             counter +=1;
        else:
            break;
    
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
movies['crew'][0] 
#In crew I only want Director as a job So below is the following method
```

```python

def fetchDirector(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break;
    return L
```

```python
movies['crew'].apply(fetchDirector)
#Here's the director list
```

```python
movies['crew'] = movies['crew'].apply(fetchDirector)
```

```python
movies.head()
```

```python
movies['overview'][0]
#As we can see overview is a string. So we have to convert it into a list
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
# Now we have to remove all the spaces
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
# Now we will merge all the necessary columns into a single column named tags
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
new_df = movies[['movie_id', 'title','tags']]  
```

```python
new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
#converting the tags list into a string
```

```python
new_df.head()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
    
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df['tags'][0]
```

```python
#converting the string into lowercase
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
#Now we will vectorize the tags by the technique bag of words
import sklearn
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words = "english")
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
#visual representation of those 5000 mostly used words
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
cosine_similarity(vectors).shape
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
        
   
```

```python
recommend('Batman Begins')
```
